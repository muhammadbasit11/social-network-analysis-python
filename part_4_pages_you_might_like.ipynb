```python
import json
def loading_data(filename):
    with open(filename,"r") as file:
        data=json.load(file)
        return data
```

```python
def pages_you_might_like(user_id,data):
    all_users_pages={}
    for users in data["users"]:
        all_users_pages[users["id"]]=set(users["liked_pages"])
    if user_id not in all_users_pages:
        return []
    user_pages=all_users_pages[user_id]
    suggested_pages={}
    for other_users,pages in all_users_pages.items():
        if user_id!=other_users:
            mutual_pages=user_pages.intersection(pages)
            if mutual_pages:
                for page in pages:
                    if page not in user_pages:
                        suggested_pages[page]=suggested_pages.get(page,0) + len(mutual_pages)  #where len tells that the more we both share pages the more other pages you will get
    sorted_suggestions=sorted(suggested_pages.items(),key=lambda x:x[1],reverse=True)
        
    return list(map(lambda x:x[0],sorted_suggestions))
```

```python
data=loading_data("massive_data.json")
user_id=2
pages_you_might_like(user_id,data)
```
