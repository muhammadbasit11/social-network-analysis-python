```python
#loading json data or file into the program
import json
```

```python
def load_file(filename):
    with open(filename,"r") as file:
        content=json.load(file)
        return content
```

```python
data=load_file("unstrctued_data.json")
# print(data)
```

```python
#main code
def cleaning_data(data):
    #removing users with no names
    data["users"]=[user for user in data["users"] if user["name"].strip()]
    #removing users with duplicate friends 
    for user in data["users"]:
        user["friends"]=list(set(user["friends"]))
    #removing users with no friends and pages:
    data["users"]=[user for user in data["users"] if user["friends"] and user["liked_pages"]]

    #NOW REMOVING THE PAGES BASED ON DUPLICATE ID'S
    unique_pages={}
    for page in data["pages"]:
        unique_pages[page["id"]]=page
    data["pages"]=list(unique_pages.values())
    return data


    #we can dump this data into a file
with open("cleaned_data.json","w") as file:
    json.dump(data,file)
```

```python
cleaning_data(data)
```
