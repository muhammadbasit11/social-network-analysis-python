```python
import json
def loading_data(filename):
    with open(filename,"r") as file:
        data=json.load(file)
        return data
def people_you_may_know(user_id,data):
    all_users_friends={}                 #it will store all the users'id and their friends
    for user in data["users"]:
        all_users_friends[user["id"]]=set(user["friends"])
    if user_id not in all_users_friends:
        return []
    
    user_friends=all_users_friends[user_id]   #it will store the friends of the user whose mutuals freinds we are looking
    friends_of_user_friends=[]
    for friends in user_friends:
        friends_of_user_friends.extend(all_users_friends[friends])#here extend will be used instead of append becuase if we use
                                                             #append it will add up lists in list but extend add the elements only
    suggestions={}
    for mutual in friends_of_user_friends:
        if user_id!=mutual and mutual not in user_friends:
            suggestions[mutual]=suggestions.get(mutual,0)+1 #first it will check whether this key exist or not if not then it creates
                                                   #that key and assign the value to it and when the same key occurs again then
                                                   #it gives the value and increment 1 to the value and assign it
                                                                
    
    sorted_suggestions=sorted(suggestions.items(),key=lambda x:x[1],reverse=True) #we will use sorted because it will sort any iterable
                                                                        #and sort only sort the list and it is sorted acc to key 
    return list(map(lambda x:x[0],sorted_suggestions))
    # return sorted_suggestions
    
```

```python
data=loading_data("massive_data.json")
user_id=3

recommendation=people_you_may_know(user_id,data)

print(f"the people you may know for user {user_id} is: {recommendation}")
```
