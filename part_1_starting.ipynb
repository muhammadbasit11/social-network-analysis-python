```python
import json
```

```python
#loading the json data 
def load_json_data(filename):
    with open(filename,"r") as file:
        data=json.load(file)
        return data
```

```python
data=load_json_data("data.json")
```

```python
#displaying the users and their connections
```

```python
#main program
def show_details(x):
    print("USERS AND THEIR CONNECTIONS:")
    #FOR THE USER DETAILS:
    for user in x["users"]:
        print(f"{user["name"]} (ID:{user["id"]})-FRIENDS:{user["friends"]}-LIKED PAGES:{user['liked_pages']} ")
    print("\nPAGES DETAILS:")
    #FOR THE PAGES DETAILS:
    for page in x["pages"]:
        print(f"{page["id"]}:{page["name"]}")
```

```python
show_details(data)
```
